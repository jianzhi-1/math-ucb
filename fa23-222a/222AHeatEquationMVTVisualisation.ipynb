```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 0
t = 0
d = 1
r = 4
r2 = 6
```

```python
def E(t, x):
    if t <= 0: return 0
    return 1./((4.*np.pi*t)**(d/2))*np.exp(-x**2/(4.*t))
```

```python
y = np.random.uniform(low=-2, high=2, size=10000)
s = np.random.uniform(low=-2, high=2, size=10000)
```

```python
accept = [(ss, yy) for (ss, yy) in zip(s, y) if E(t - ss, x - yy) >= (1./r)**d]
reject = [(ss, yy) for (ss, yy) in zip(s, y) if E(t - ss, x - yy) < (1./r)**d]
```

```python
plt.scatter([x[1] for x in accept], [x[0] for x in accept], label='accept', c='green')
plt.scatter([x[1] for x in reject], [x[0] for x in reject], label='reject', c='red')
plt.scatter([0],[0],s=100)
plt.legend()
plt.show()
```

```python
accept2 = [(ss, yy) for (ss, yy) in zip(s, y) if E(t - ss, x - yy) >= (1./r2)**d]
reject2 = [(ss, yy) for (ss, yy) in zip(s, y) if E(t - ss, x - yy) < (1./r2)**d]
```

```python
plt.scatter([x[1] for x in accept2], [x[0] for x in accept2], label='accept (r = {})'.format(r2), c='green')
plt.scatter([x[1] for x in reject2], [x[0] for x in reject2], label='reject', c='red')
plt.scatter([x[1] for x in accept], [x[0] for x in accept], label='accept (r = {})'.format(r), c='lime')
plt.scatter([0],[0],s=100)
plt.legend()
plt.show()
```
